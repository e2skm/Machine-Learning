# Linear Regression - Canada annual IPC dataset - ML project9

## Import the necessary liabraries

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Load dataset into pandas dataframe 

```python
df = pd.read_csv(r'C:\Users\itumeleng\Desktop\Python Programs\Canada_per_capita_income.csv')
df.info()
```

```python
df.head()
```

## Visualize the data distribution and regression line

```python
sns.regplot(x='year',y='income',data=df)
```

## Create and Train model

```python
x = df[['year']].values
y = df['income']
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

## Make a prediction (for the year 2021)

```python
prediction2021 = model.predict([[2021]])
print(f'The predicted IPC in Canada for 2021 is: {prediction2021}')
```

## Evaluate the model:
### The actual IPC in Canada for the year 2021 is $ 41979

```python
error = 42117 - 41979
print('Our model predicted $',error,'extra which is quite impressive!!')
```

### As seen our model is really accurate, since it made a close prediction
