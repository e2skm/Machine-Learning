## Data preprocessing: prepare the data for the model

### Split the data into Predictor (X) and target (y) varibles respectively

```python
X = whiteWineQuality_df.drop('quality',axis=1)
```

```python
y = whiteWineQuality_df['quality']
```

### Scale the predictor variables

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Split X and Y datasets into train and test sets respectively

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.20,random_state=42)
```

## Implement an ML model

### Initialize the model

```python
model = LinearRegression()
```

### Train the model

```python
model.fit(X_train,y_train)
```

### Used the trained model to make predictions using the remaining predictor variables

```python
y_pred = model.predict(X_test)
for i in range(len(y_pred)):
    y_pred[i] = int(round(y_pred[i]))
```

## Evaluate the model: Compare the model's predictions to actual target outputs

### Get the accuracy score 

```python
accuracy = accuracy_score(y_pred,y_test) * 100
print(f'This model has an accuracy of {round(accuracy)} % in predicting the quality of white wine')
```

### Get the classification report

```python
classRep = classification_report(y_pred,y_test)
classRep
```

### Get the confusion matrix

```python
confusionMat = confusion_matrix(y_pred,y_test)
```

```python
sns.heatmap(confusionMat, annot=True,fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Output')
plt.ylabel('Actual Output')
plt.show()
```
