# Three types of Naive Bayes classifier with Scikit learn - Iris dataset - ML project8

## 1. Gaussian Naive Bayes

### Import the necessary liabraries

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
```

### Load the dataset 

```python
dataset = datasets.load_iris()
```

### Create and train model

```python
modelGNB = GaussianNB()
```

```python
modelGNB.fit(dataset.data,dataset.target)
```

### Make Predictions

```python
expected = dataset.target
predictedGNB = modelGNB.predict(dataset.data)
```

### Evaluate the model

#### Get Confusion matrix 

```python
confuMat = metrics.confusion_matrix(expected,predictedGNB)
confuMat
```

#### Get Classification report

```python
classRep = metrics.classification_report(expected,predictedGNB)
classRep
```

#### Get accuracy

```python
accuracyGNB = round((metrics.accuracy_score(expected,predictedGNB) *100),2)
print('The accuracy of the gaussian modedl is: {0}%'.format(accuracyGNB))
```
